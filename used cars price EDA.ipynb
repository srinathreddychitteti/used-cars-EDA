# Cars price dataset Exploratory Data Analysis

## Goal of the project:
The final goal of the project is to understand the data and make visualistions about the dataset

### The Dataset:
The dataset is taken from kaggle it contains 852122 rows of data and 8 distinct columns
The dataset provides data of used cars listed in Craigslist.com. All of the listings are from various
states from the country of USA

### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_rows', 6000)
```

```python
data_path = "./true_car_listings.csv"
```

```python
org_data = pd.read_csv(data_path)
```

```python
org_data.head()
```

```python
org_data.shape
```

### Describing the type of data

```python
org_data.info()
```

Unique cities from which the listings are made

```python
pd.value_counts(org_data["City"])
```

It seems that most of the listings are made from the city of Houston

```python
len(pd.unique(org_data["City"]))
```

Unique states from which the listings are made

```python
pd.value_counts(org_data["State"])
```

It seems that most of the listings are made from the state of Texas

```python
len(pd.unique(org_data["State"]))
```

Bar chart depecting the number of listings per city

```python
fig, ax = plt.subplots(figsize=(20, 6))
city_plot = sns.barplot(x = pd.unique(org_data["City"]) , y = pd.value_counts(org_data["City"]) , data = org_data)
city_plot.set(xlabel = "Cities" , ylabel = "Count")
city_plot.tick_params(axis='x', rotation=90)
```

Bar chart depecting the number of listings per State

```python
fig, ax = plt.subplots(figsize=(20, 6))
state_plot = sns.barplot(x = pd.unique(org_data["State"]) , y = pd.value_counts(org_data["State"]) , data = org_data)
state_plot.set(xlabel = "States" , ylabel = "Count")
state_plot
```

Unique "make" from which the listings are made

```python
pd.value_counts(org_data["Make"])
```

It seems that cars made by ford company are listed

Bar chart depecting the number of listings per "Make"

```python
fig, ax = plt.subplots(figsize=(20, 6))
make_plot = sns.barplot(x = pd.unique(org_data["Make"]) , y = pd.value_counts(org_data["Make"]) , data = org_data)
make_plot.set(xlabel = "Make" , ylabel = "Count")
make_plot.tick_params(axis='x', rotation=90)
```

Unique "Models" from which the listings are made

```python
pd.value_counts(org_data["Model"])
```

It seems that Silverado is the highest listed car

```python
fig, ax = plt.subplots(figsize=(20, 6))
make_plot = sns.barplot(x = pd.unique(org_data["Model"]) , y = pd.value_counts(org_data["Model"]) , data = org_data)
make_plot.set(xlabel = "Model" , ylabel = "Model")
make_plot.tick_params(axis='x', rotation=90)
```
